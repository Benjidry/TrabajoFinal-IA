```python
import pandas as pd
```

```python
labelMap = pd.DataFrame()
listDataType = []

stringType = ""
stringNumber = ""

for type in range(0,4):
    if type == 0:
        stringType = "H"
    elif type == 1:
        stringType = "S"
    elif type == 2:
        stringType = "D"
    elif type == 3:
        stringType = "C"
    for number in range(0,13):
        if number == 0:
            stringNumber = "A"
        elif number > 0 and number < 10:
            stringNumber = str(number + 1)
        elif number == 10:
            stringNumber = "J"
        elif number == 11:
            stringNumber = "Q"
        elif number == 12:
            stringNumber = "K"

        dataKey = stringNumber + stringType
        
        listDataType.append(dataKey)

labelMap = pd.DataFrame(listDataType)
labelMap = labelMap.rename({0: "name:"}, axis="columns")
labelMap["id:"] = labelMap.index + 1
labelMap
```

```python
labelMap.to_json('label_map.pbtxt', orient= "records")
```
