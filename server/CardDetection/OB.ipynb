```python
from ultralytics import YOLO
import os 
import numpy as np 
```

```python
modelCard = YOLO("../last.pt")
```

```python
def detectCard(model, path):
    results = model([path])
    print(results)

    for result in results:
        boxes = result.boxes
        for box in boxes:
            print(box)
        masks = result.masks
        keypoints = result.keypoints
        probs = result.probs
        obb = result.obb
        names = result.names
        result.save(filename=path+"OD.jpg")
    
    return boxes, names
```

```python
def parseBoxes(boxes, names):
    cards = []
    
    if len(boxes) == 0:
        return cards

    cardsDealer = []
    cardsPlayer = []
    
    orig_shapeX, orig_shapeY = boxes[0].orig_shape
    
    print(orig_shapeX)
    print(orig_shapeY)
    
    #It is going to return the type in which it is being considered and the type 
    
    print(names)

    for box in boxes:
        print("Type Card")
        TypeCard = names[int(box.cls.item())]
        print(TypeCard)

        print("Confidence")
        Confidence = round(box.conf.item(),2)
        print(Confidence)
        
        #Card Dictionary
        card = {
            "Conf": Confidence,
            "TypeCard": TypeCard
        }
        
        #From a box 
        #X1Y1
        #       X2Y2
        print("Coordinates")
        #It prints like (X1) 
        print(box.xyxy.numpy())
        Coordinates = box.xyxy.numpy()[0]
        X1 = Coordinates[0]
        Y1 = Coordinates[1] 
        X2 = Coordinates[2]
        Y2 = Coordinates[3]
        
        if Y1 > orig_shapeX / 2 or Y2 > orig_shapeX / 2:
            cardsDealer.append(card)
        else: 
            cardsPlayer.append(card)
        
        print(X1, X2, Y1, Y2)
        
    return cardsDealer, cardsPlayer
```

```python
boxes, names = detectCard(modelCard, "../screenshots/playing-cards-aces-set_78370-2354.png")
```

```python
cardDealer, cardPlayer = parseBoxes(boxes, names)
```

```python
def parseCards(setCards):
    cardsParsed = []
    
    for card in setCards:
        toParse = True
        
        if len(cardsParsed) == 0:
            toParse = True
                
        else:
            for cardP in cardsParsed:
                if cardP["TypeCard"] == card["TypeCard"]:
                    toParse = False 
                    if card["Conf"] > cardP["Conf"]:
                        cardP["Conf"] = card["Conf"]
            
        if toParse:    
            cardsParsed.append(card) 
           
    print(cardsParsed)
    return cardsParsed
```

```python
print(cardDealer)
print(cardPlayer)
```

```python
dealerParsed = parseCards(cardDealer) 
playerParsed = parseCards(cardPlayer)
```

```python
def translateArray(cards):
    array = np.zeros(13)
    
    for card in cards:
        print(card["TypeCard"])
        cat = categoryCard(card["TypeCard"])
        value = valueCard(card["TypeCard"], cat)
        array[value - 1] = array[value - 1] + 1

    print(array)
    return array


typesCards = ["S", "D", "C", "H"]


def categoryCard(card):
    if typesCards[0] in card:
        return 0
    elif typesCards[1] in card:
        return 1
    elif typesCards[2] in card:
        return 2
    elif typesCards[3] in card:
        return 3


def valueCard(card, cat):
    value = card[0].split(typesCards[cat])[0]
    if value == "A":
        return 1
    elif value == "J":
        return 11
    elif value == "Q":
        return 12
    elif value == "K":
        return 13
    else:
        return int(value)


def countCards(arrayCards):
    arrayTMP = arrayCards
    count = 0
    ace = False
    for index in range(13):
        if arrayTMP[index] >= 1:
            if index == 0:
                ace = True
                if arrayTMP[index] == 2:
                    count = 21
                    ace = False
                else:
                    count = arrayTMP[index] * 1 + count
            elif index == 10:
                count = count + arrayTMP[index] * 10
            elif index == 11:
                count = count + arrayTMP[index] * 10
            elif index == 12:
                count = count + arrayTMP[index] * 10
            else:
                count = count + arrayTMP[index] * (index + 1)

    if ace and count < 11:
        count = count + 11

    return count-1
```

```python
arrayDealer = translateArray(dealerParsed)
arrayPlayer = translateArray(playerParsed)
```

```python
print(countCards(arrayDealer))
print(countCards(arrayPlayer))
```
